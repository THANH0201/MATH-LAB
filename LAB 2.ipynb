```python
"""
PROBLEM 4: Using linearly spaced numerical array of x from 0 to 10 having 100 samples draw in the same graph two polynomials: y1 = (x-2).(x-8) y2 = (x-2).(x-5).(x-8)

"""
x=np.linspace(0,10,100)
y1=(x-2)*(x-8)
y2=(x-2)*(x-5)*(x-8)

plt.plot(x, y1)
plt.plot(x, y2)
plt.grid(True)
plt.show()
```

```python
"""
PROBLEM 4: Using linearly spaced numerical array of x from 0 to 10 having 100 samples draw in the same graph two polynomials: y1 = (x-2).(x-8) y2 = (x-2).(x-5).(x-8) Add labels 'y1' and 'y2' to the plots. Add grid lines to the graph. Add x-label: 'x-axis'. Add y-label: 'y-axis'. Add title: 'Polynomial functions'. Add legend to the the graph.
"""
x=np.linspace(0,10,100)
y1=(x-2)*(x-8)
y2=(x-2)*(x-5)*(x-8)

plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Polynomial function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
"""
PROBLEM 5: Using linearly spaced numerical array of x from 0 to 10 having 100 samples calculate the values for the two polynomials: y1 = (x-2).(x-8) y2 = (x-2).(x-5).(x-8) Using two subplots, plot the first polynomial on the first axes and then the second polynomial on the second axes.

Use figure size of (6, 8). Annotate the graph:

Add grid lines to both plots. Add title: 'y1' to the first subplot. Add title: 'y2' to the second subplot. Add x-label: 'x-axis' to the second plot.
"""
x=np.linspace(0,10,100)
y1=(x-2)*(x-8)
y2=(x-2)*(x-5)*(x-8)
fig, ax = plt.subplots(2, figure = (6,8))
plt.sca(ax[0])
plt.plot(x, y1, label = "y1")
plt.grid(True)
plt.legend()

plt.sca(ax[1])
plt.plot(x, y2, label = "y2")
plt.grid(True)
plt.xlabel('x-axis')
plt.legend()
plt.show()
```

```python
"""
PROBLEM 6: Calculate values for two functions f(x) = and g(x) = 5 - x . Use linearly spaced numerical array of x from 0 to 10 having 100 samples. Use two subplots and figure size of (6, 8). Draw the full graphs of both functions into the first axes. This part of the graph shows the full graphs. Draw the full graphs of both functions also into the second axes. This part is used for zooming in to the details. Add grid lines to both subplots. Change the x-axis and y-axis limits for the second axes, so that you can find graphically the intersection point (x0, y0) for the two equations (e. f(x) = g(x) when ). Try to get the value minimum three significant figures accuracy. Read the intersection point (x0, y0) from the graph and draw it to the second graph using red circle.
"""
x=np.linspace(0,10,100)
f = np.sqrt(2*x)
g = 5 - x

fig, ax = plt.subplots(2, figure = (6,8))
plt.sca(ax[0])
plt.plot(x, f, label = "f(x)", color = "r")
plt.plot(x, g, label = "g(x)", color = "b")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('FULL GRAPHS')
plt.grid(True)
plt.legend()

plt.sca(ax[1])
plt.xlim(2.67,2.69)
plt.ylim(2.31,2.33)
plt.plot(x, f, label = "f(x)", color = "r")
plt.plot(x, g, label = "g(x)", color = "b")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('DETAIL GRAPHS')
plt.grid(True)
plt.legend()
plt.show()
# Intersection point
x0 = 2.683
y0 = 2.316
plt.plot(x0, y0, "o", color = "r")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('DETAIL GRAPHS')
plt.grid(True)
plt.legend()
plt.show()
```
